```python
import pandas as pd
d=pd.read_csv("/content/Salary Data.csv")
```

```python
import numpy as np
d.replace(0,np.nan)
d=d.dropna()
d.isna().sum()
```

```python
from pycaret.regression import *
s=setup(d,target='Salary')
```

```python
best=compare_models()
```

```python
final_model=finalize_model(best)
pred=predict_model(final_model,data=d)
pred
```
